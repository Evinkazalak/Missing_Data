# EKSİK VERİLER

```python
import pandas as pd
import numpy as np
```

```python
sozlukVerısı={"ıstanbul":[30,29,np.nan],"ankara":[20,np.nan,25],"ızmır":[23,45,67]}
havaDurumuDataFrame=pd.DataFrame(sozlukVerısı)
```

```python
havaDurumuDataFrame
```

```python
#nan yazan yerler için farklı çözümler var. ilki şudur : olmayan verileri komple sil bunun için dropna komutunu kullan
#bu komut içinde veri olmayan kısımları direkt sil demektir. ve eğer satırda nan varsa satırı komple sil.
havaDurumuDataFrame.dropna()
```

```python
#kolonu silmek istersen eğer axix değerini 1 ver
havaDurumuDataFrame.dropna(axis=1)
```

```python
#örn 2 tane nan olanları drop  et ama tek olanları drop etme
YenıVeri={"ıstanbul":[30,29,np.nan],"ankara":[20,np.nan,25],"ızmır":[23,45,67],"antalya":[35,np.nan,np.nan]}
YenıDataFrame=pd.DataFrame(YenıVeri)
```

```python
YenıDataFrame
```

```python
#örn 2 tane nan olanları drop  et ama tek olanları drop etme
# bunun için thresh komutunu kullan

YenıDataFrame.dropna(axis=1,thresh=2)
```

```python
#boş verilere standart bir sayı eklemek istersem eğer ne olacak?
YenıDataFrame.fillna(34)
```

# GURUPLANDIRMA

```python
maasSozlugu={"departman":["yazılım","pazarlama","satış","finans","insan kaynakları","hukuk"]
           ,"calısan ismi":["süleyman","halil","mehmet","metin","sibel","meryem"],"maas":[100,200,300,400,500,700]}
```

```python
maasDataFrame=pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
#group by fonksiyonu ben neye göre guruplandırmak istiyorsam ağer ona göre guruplandırma yapıyor.
grupObjesi=maasDataFrame.groupby("departman")
```

```python
grupObjesi.count()
```

```python
# mean : ortalama hesaplıyor
grupObjesi.mean()
```

```python
grupObjesi.max()
```

```python
grupObjesi.min()
```

```python
#describe:  her türlü bilgiyi bana veriyor
grupObjesi.describe()
```
